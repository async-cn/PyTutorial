# J5-B 折扣方案

因为 Maaike 太过可爱，Louis 给 Maaike 提供了三种折扣方案。

- 方案 Ⅰ：打 $8$ 折。
- 方案 Ⅱ：打 $6$ 折，然后增加 $15$ 元。
- 方案 Ⅲ：增加 $25$ 元，然后打 $5$ 折。

由于 Maaike 不会算数，请你帮助 Maaike 计算三种折扣方案分别打折后的价格。

## 输入格式

第一行一个**浮点数** $p$。

## 输出格式

三行，每行一个浮点数，分别表示三种折扣方案打折后的价格 $p_1,p_2,p_3$，保留 $2$ 位小数。

## 输入输出样例

### 样例输入 #1

```text
100
```

### 样例输出 #1

```text
80.00
75.00
62.50
```
