# J7-A 闰年判断

人们在使用历法时发现，随着时间的积累，历法所体现的日历日期与实际季节之间的总会渐行渐远，产生较大误差。因此，为了确保日期与季节之间的对应关系不断保持。人们通过引入闰年的方式，调整日历，使其与地球公转周期保持一致。现行公历的置润规则：**年份能被 $4$ 整除且不能被 $100$ 整除，或能被 $400$ 整除的年份为闰年。**

请你制作一个 Python 程序，输入年份 $n$，判断其是否是闰年。

## 输入格式

第一行一个整数 $n$。

## 输出格式

一行，一个字符串，若是闰年，则为 `YES`；若不是闰年，则为 `NO`。

## 输入输出样例

### 样例输入 #1

```text
2020
```

### 样例输出 #1

```text
YES
```

### 样例输入 #2

```text
3000
```

### 样例输出 #2

```text
NO
```

## 说明与提示

使用模运算可以进行整除判断。例如以下条件可以判断 $n$ 是否能被 $k$ 整除：

```python
n % k == 0
```

对于所有测试数据，$0\le n\le {10}^9$
