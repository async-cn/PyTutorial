## 比较运算符

在 Python 中，使用比较运算符可以比较两个值之间的关系。

以下两种运算符几乎适用于所有 Python 数据类型：

| 比较运算符 | 名称  | 作用  | 真值条件 |
|-------|-----|-----|------|
| `==`  | 等于  | 判断两个值是否相等 | `==` 符号左右两侧的值相等 |
| `!=`  | 不等于 | 判断两个值是否不相等 | `!=` 符号左右两侧的值不相等 |

**注意：单独一个 `=` 是赋值运算，连续两个 `=` 组成的 `==` 才是“等于”比较运算符！**

以下运算符适用于*数值*（整数或浮点数）之间的比较：

| 比较运算符 | 名称 | 真值条件       |
| --- | --- |------------|
| `>` | 大于 | 左侧数值大于右侧数值 |
| `<` | 小于 | 左侧数值小于右侧数值 |
| `>=` | 大于 | 左侧数值大于或等于右侧数值 |
| `<=` | 小于 | 左侧数值小于或等于右侧数值 |

注意：由于*浮点数精度限制*，两个浮点数之间不建议使用`==`、`>=`或`<=`直接进行比较，而是应**取近似值**后再进行比较。

### 试一试

````python
n = int(input("请输入整数: "))

if n > 10:
    print("n 大于 10 成立。")

if n >= 10:
    print("n 大于等于 10 成立。")

if n == 15:
    print("n 等于 15 成立。")
````

在 Python 中，同方向的数值比较运算符可以在同一表达式内连续使用，以判断数值是否在指定区间内。

这种语法在其他编程语言中通常不被支持，而是需要使用*逻辑与*运算（`and` 或 `&&`）手动实现，这种 Python 支持但其他编程语言很少支持的便捷语法称为 *语法糖*。

````python
n = int(input("请输入整数: "))
if 5 < n < 10:
    print("n in (5, 10)")
if 5 <= n < 10:
    print("n in [5, 10)")
if 5 < n <= 10:
    print("n in (5, 10]")
if 5 <= n <= 10:
    print("n in [5, 10]")
````

## 逻辑运算

在 Python 中，逻辑运算符可以进行复杂的逻辑运算。

常用的逻辑运算符：

| 逻辑运算符 | 名称 | 作用           | 语法 |
| --- | --- |--------------| --- |
| `and` | 与 | 两侧布尔值均为真     | `a and b` |
| `or` | 或 | 其中一侧或两侧布尔值为真 | `a or b` |
| `not` | 非 | 反转真值 | `not a` |

其中 `and` 和 `or` 为二元运算，`not` 为一元运算。

以下是三个常用逻辑运算符的*真值表*：（✅ 代表 `True`，❌️ 代表 `False`）

### 与

表达式：`a and b`

| `a`的值 | `b`的值 | `a and b` 的值 |
|-------| --- |--------------|
| ✅     | ✅ | ✅            |
| ✅     | ❌️ | ❌️           |
| ❌️    | ✅ | ❌️           |
| ❌️    | ❌️ | ❌️           |

### 或

表达式：`a or b`

| `a`的值 | `b`的值 | `a or b` 的值 |
|-------| --- |------------|
| ✅     | ✅ | ✅          |
| ✅     | ❌️ | ✅          |
| ❌️      | ✅ | ✅          |
| ❌️    | ❌️ | ❌️         |

### 非

表达式：`not a`

| `a`的值 | `not a` 的值  |
|-------|-------------|
| ✅ | ❌️          |
| ❌️ | ✅           |

### 试一试

输入不同的真值，看看会有什么不同的输出。

````python
def show(exp, x): # 用于展示 x 的真值并附带好看的颜色
    print(exp, "的值为:", "\033[92mTrue\033[0m" if x else "\033[91mFalse\033[0m")

a = input("请输入 a 的真值（T代表True, F代表False）: ") == 'T'
b = input("请输入 b 的真值（T代表True, F代表False）: ") == 'T'

show('a', a)
show('b', b)

print() # 输出一个空行

show("a and b", a and b)
show("a or b", a or b)

print() # 输出一个空行

show("not a", not a)
show("not b", not b)
````

### 试一试

有一种二元逻辑运算，叫做“异或”（xor），当且仅当两个输入值**不同**时其运算结果才为 `True`，其真值表如下：

| `a`的值 | `b`的值 | `a xor b` 的值 |
| --- | --- |--------------|
| `True` | `True` | `False`      |
| `True` | `False` | `True`       |
| `False` | `True` | `True`       |
| `False` | `False` | `False`      |

令人遗憾的是，Python 里没有这种神奇的逻辑运算。作为一名有点实力的 Python 程序员，这怎么能忍？

请制作一个 Python 程序，输入 `a` 和 `b` 的真值，将变量 `c` 的值设为 `a xor b` 的结果并输出。

> 小技巧：使用括号可以改变逻辑运算的优先级，括号内的逻辑表达式会先被计算。

````python
# It's your turn!
````

**答案**

根据定义，a与b的真值不同等价于 *a 与 非b 都为`True`* 或 *b 与 非a 都为`True`*。

因此，组合后的表达式如下：

````python
c = (a and not b) or (b and not a)
````

注意到 `xor` 的真值与 `or` 的真值只有在 `a` 和 `b` 均为 `True` 时不同，即 `a and b` 为 `False` 时 `a xor b` 的值为 `a or b` 的值，`a and b` 为 `True` 时 `a xor b` 的值为 `False`。

因此也可以这样实现：

````python
c = (a or b) and not (a and b)
````
