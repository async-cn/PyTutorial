# J8-D 下一个闰年

众所周知，年份能被 $4$ 整除且不能被 $100$ 整除，或能被 $400$ 整除的年份为闰年。在中国，每个闰年的二月份都比原来多出一天，即由 $28$ 天增加至 $29$ 天。

作为高中生的 Maaike 很喜欢闰年，因为 Maaike 的学校刚好在每年的 $3$ 月 $1$ 日开学，所以每个闰年 Maaike 的寒假都能延长一天。

可惜的是，并不是每一年都是闰年，所以 Maaike 很想知道下一个闰年的年份。给定当前年份 $n$，请你帮 Maaike 找到下一个闰年，即最小的年份 $x$ 满足 $x>n$ 且 $x$ 是闰年。

## 输入格式

一行，一个整数 $n$，表示当前年份。

## 输出格式

一行，一个整数，表示下一个闰年的年份。

## 输入输出样例

### 样例输入 #1

```text
2010
```

### 样例输出 #1

```text
2012
```

### 样例输入 #2

```text
2020
```

### 样例输出 #2

```text
2024
```

## 说明与提示

对于所有测试数据，$0\le n\le {10}^9$
