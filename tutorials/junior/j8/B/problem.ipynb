# J8-B 绿水晶与蓝水晶

Maaike 又遇到了一个生长着许多水晶的洞穴，特别的是，这个洞穴中的水晶有着绿和蓝两种颜色，更加美丽。

洞穴内的水晶们成簇生长，且从洞穴入口开始，所有水晶簇共同排成了一条直线。设距离洞口最近的水晶簇是第 $1$ 个水晶簇，经过观察，Maaike 发现第 $1$ 个水晶簇中有 $1$ 个绿水晶，第 $2$ 个水晶簇中有 $2$ 个蓝水晶 $\cdots$，以此类推，第 $i$ 个水晶簇中有 $i$ 个水晶，且序号为奇数的水晶簇中水晶全部为绿色，序号为偶数的水晶簇中水晶全部为蓝色。

洞穴内一共有 $n$ 簇水晶，请你帮助 Maaike 分别计算绿水晶和蓝水晶的总数量。

## 输入格式

一行，一个整数，为水晶簇总数 $n$。

## 输出格式

一行，两个整数（以空格分隔）$s_1,s_2$，依次表示绿水晶和蓝水晶的总数量。

## 输入输出样例

### 样例输入 #1

```
6
```

### 样例输出 #1

```text
9 12
```

### 样例输入 #2

```text
100
```

### 样例输出 #2

```text
2500 2550
```

## 样例解释

对于样例 #1：
- 绿水晶数量：$1+3+5=9$
- 蓝水晶数量：$2+4+6=12$

对于样例 #2：
- 绿水晶数量：$1+3+\cdots+97+99=2500$
- 蓝水晶数量：$2+4+\cdots+98+100=2550$

## 说明与提示

对于所有测试数据，$0\le n\le 65536$
