# J8-C 可爱之力

Maaike 是猫咪军队的主帅，他军队中共有 $n$ 只猫咪，它们的可爱值分别为 $a_1,a_2,\cdots,a_n$，且每只猫咪的可爱值与自身的编号相等，即所有猫咪均满足 $a_i=i$。

猫咪军队的战力由军队的 “可爱之力” $s$ 决定。可爱之力 $s$ 初始为 $0$，在开始战斗之后，所有猫咪会相互握手，每当编号为 $i$ 和 $j$ 的两只猫咪握手时，若本次战斗中这两只猫咪之前没有相互握过手，可爱之力 $s$ 便会增加 $a_i\times a_j$。直到所有猫咪都相互握手之后，可爱之力 $s$ 才会达到最大值 $s_{max}$。

在出征之前，Maaike 需要先评估军队的最强战斗力。请你帮助 Maaike 计算军队的可爱之力 $s$ 的最大值 $s_{max}$。

## 输入格式

一行，一个整数，为猫咪总数 $n$。

## 输出格式

一行，一个整数，为可爱之力最大值 $s_{max}$。

## 输入输出样例

### 样例输入 #1

```
4
```

### 样例输出 #1

```text
35
```

### 样例输入 #2

```text
100
```

### 样例输出 #2

```text
12582075
```

## 样例解释

对于样例 #1，$s_{max} = (1\times 2)+(1\times 3)+(1\times 4)+(2\times 3)+(2\times 4)+(3\times 4) = 35$

<img src="images/img-1.png" width="25%">

对于样例 #2：$s_{max} = \sum_{x=2}^{100}\sum_{y=1}^{x-1} xy = 12582075$

## 说明与提示

对于所有测试数据，$2\le n\le 200$
