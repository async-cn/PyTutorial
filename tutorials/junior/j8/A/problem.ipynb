# J8-A 水晶统计

Maaike 在一处洞穴中遇到了许多水晶，它们非常稀有且美丽。

洞穴内的水晶们成簇生长，且从洞穴入口开始，所有水晶簇共同排成了一条直线。设距离洞口最近的水晶簇是第 $1$ 个水晶簇，经过观察，Maaike 发现第 $1$ 个水晶簇中有 $1$ 个水晶，第 $2$ 个水晶簇中有 $2$ 个水晶 $\cdots$，以此类推，第 $i$ 个水晶簇中有 $i$ 个水晶。

洞穴内一共有 $n$ 簇水晶，请你帮助 Maaike 计算水晶的总数量。

## 输入格式

一行，一个整数，为水晶簇总数 $n$。

## 输出格式

一行，一个整数，表示洞穴内的水晶总数。

## 输入输出样例

### 样例输入 #1

```
5
```

### 样例输出 #1

```text
15
```

### 样例输入 #2

```text
100
```

### 样例输出 #2

```text
5050
```

## 样例解释

对于样例 #1，水晶数量：$1+2+3+4+5=15$

对于样例 #2，水晶数量：$1+2+\cdots+99+100=5050$

## 说明与提示

对于所有测试数据，$0\le n\le 65536$
