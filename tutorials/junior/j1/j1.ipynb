# J1 你好，世界！——基本输出

print 函数是最常用的 Python 内置函数之一，用于在控制台打印输出。

试试以下代码：

```python
print("Hello, World!")
```

```python
print("Hello, Python!")
print("Hello, Java!")
print("Hello, C++!")
```

是不是很神奇呀！
注意，使用 print 直接输出文本时，需要加上双引号，以标记输出的内容为*字符串*。

### 试一试
使用 print 语句输出 “你好，世界！” （在下方输入代码并运行）

使用逗号分割多个参数即可在一行内输出多个内容，他们将以空格分隔。

```python
print("Hello","Louis")
print("Days in each week:","Mon","Tue","Wed","Thu","Fri","Sat","Sun")
```

print 可以自动将输入的参数全部转换为字符串，无需考虑类型问题喵！

```python
print("中奖号码:",114,514,191,981) # 整数
```

```python
print("所有商品价格:", 9.1, 114.5, 20.26, 100) # 浮点数
```

```python
print("已购买会员:",False) # 布尔值-False
print("未购买会员:",True) # 布尔值-True
```
