## 整数和浮点数

整数即整数，浮点数即小数。

二者区别不大，只有少数场景中需要进行区分。大多数能使用float的情况下都能使用int，但有些只能使用int的情况下不能使用float。

使用 int 和 float 函数可以将对应参数转换为整数和浮点数。

整数和浮点数都能进行加（+）、减（-）、乘（*）、除（/）、整除（//）、幂（**）、模（%）运算。

整数可以进行*位运算*，而浮点数通常不能。

由于*浮点数的精度限制*，不推荐使用朴素的 “`==`” 比较两个浮点数是否相等，而是应该取近似值进行比较。

```python
a = 114
b = 5.14

num1 = "100"
num2 = "20.26"
x = int(num1)
y = float(num2)

print(x, type(x))
print(y, type(y))
```

## 字符串

顾名思义，字符串就是若干个（可以为0个）字符组成的串。

使用 `len` 函数获取字符串的长度。

```python
hello = "Hello, World!"
length = len(hello)
print(length)
```

字符串直接可以使用加号（`+`）进行连接。

与print参数中使用逗号分隔的效果不同，使用加号连接字符串时不会产生额外的空格。

```python
prefix = "Mr."
firstname = "John"
lastname = "Smith"

print("Hello," + prefix + firstname + lastname) # 使用加号连接
print("Hello,", prefix, firstname, lastname) # 在 print 的参数中使用逗号分隔
```

## 布尔值

布尔值有且仅有两种合法取值：
- `True`：即“是”、“Yes”，等同于`1`
- `False`：即“否”、“No”，等同于`0`

```python
is_active = True
print(is_active, int(is_active))

is_active = False
print(is_active, int(is_active))
```
