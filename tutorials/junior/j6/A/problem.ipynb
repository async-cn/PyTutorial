# J6-A 猫猫可爱大赛

小猫 Maaike 举办了猫猫可爱大赛，但是他仅允许年龄满 $5$ 周岁且不超过 $10$ 周岁的猫猫参加。

请你制作一个 Python 程序，输入参赛猫猫的**出生年份** $n$，判断该猫猫是否满足参赛年龄限制。

具体地，给定正整数 $n$，判断 $n$ 是否满足 $2026-n\in [5, 10]$。

## 输入格式

第一行一个整数 $n$。

## 输出格式

一行，一个字符串，若满足年龄限制，则为 `YES`；若不满足，则为 `NO`。

## 输入输出样例

### 样例输入 #1

```text
2020
```

### 样例输出 #1

```text
YES
```

### 样例输入 #2

```text
2022
```

### 样例输出 #2

```text
NO
```

## 说明与提示

假如现在是 2026 年。

注意大写。

以下**条件**可以判断变量 $age$ 是否满足 $5\le age\le 10$：

```python
5 <= age <= 10
```

对于所有测试数据，$0\le n\le 2026$
