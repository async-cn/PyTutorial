# J6-C 四色问题

Maaike 偶然在网上了解到了 [四色问题](https://baike.baidu.com/item/%E5%9B%9B%E8%89%B2%E5%AE%9A%E7%90%86/805159)，但由于原版问题太难了，所以 Maaike 灵机一动，提出了自己的“四色问题”（**四** 种颜 **色** 的数字的 **问题**），规则如下：

所有正整数 $n$ 都可以被分为下列四类数字中的一种：

- 若 $n$ 同时是 $2$ 和 $3$ 的倍数，则 $n$ 是*红色数*（RED）
- 若 $n$ 是 $2$ 的倍数，但不是 $3$ 的倍数，则 $n$ 是*绿色数*（GREEN）
- 若 $n$ 不是 $2$ 的倍数，但是 $3$ 的倍数，则 $n$ 是*蓝色数*（BLUE）
- 若 $n$ 既不是 $2$ 的倍数，也不是 $3$ 的倍数，则 $n$ 是*紫色数*（PURPLE）

请你根据 Maaike 输入的数字 $n$，判断它属于四种类别中的哪一类。

## 输入格式

一行，一个整数 $n$

## 输出格式

一行，一个字符串，表示 $n$ 的分类：
- 若 $n$ 是*红色数*，则应为 `RED`
- 若 $n$ 是*绿色数*，则应为 `GREEN`
- 若 $n$ 是*蓝色数*，则应为 `BLUE`
- 若 $n$ 是*紫色数*，则应为 `PURPLE`

## 输入输出样例

### 样例输入 #1

```text
9
```

### 样例输出 #1

```text
GREEN
```

### 样例输入 #2

```text
35
```

### 样例输出 #2

```text
PURPLE
```

## 说明与提示

对于所有测试数据，$1\le n\le {10}^9$
